## O datasetu
Podaci su dobiveni istraživanjem učenika matematike u dvije portugalske srednje škole. Sadrži mnoštvo zanimljivih društvenih, rodnih i studijskih informacija o studentima. Podaci su prikupljani na način da se iz različitih kuteva pokuša utvrditi koji faktori najviše utječu na stopu konzumiranja alkohola i alkoholizma među učenicima srednjih škola.
### Sadržaj
Atributi za skup podataka student-mat.csv (učenici matematike):
1. school - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
2. sex - student's sex (binary: 'F' - female or 'M' - male)
3. age - student's age (numeric: from 15 to 22)
4. address - student's home address type (binary: 'U' - urban or 'R' - rural)
5. famsize - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
6. Pstatus - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
7. Medu - mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
8. Fedu - father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
9. Mjob - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
10. Fjob - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
11. reason - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
12. guardian - student's guardian (nominal: 'mother', 'father' or 'other')
13. traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
14. studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
15. failures - number of past class failures (numeric: n if 1<=n<3, else 4)
16. schoolsup - extra educational support (binary: yes or no)
17. famsup - family educational support (binary: yes or no)
18. paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
19. activities - extra-curricular activities (binary: yes or no)
20. nursery - attended nursery school (binary: yes or no)
21. higher - wants to take higher education (binary: yes or no)
22. internet - Internet access at home (binary: yes or no)
23. romantic - with a romantic relationship (binary: yes or no)
24. famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
25. freetime - free time after school (numeric: from 1 - very low to 5 - very high)
26. goout - going out with friends (numeric: from 1 - very low to 5 - very high)
27. Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
28. Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
29. health - current health status (numeric: from 1 - very bad to 5 - very good)
30. absences - number of school absences (numeric: from 0 to 93)

Oznake za ocjene (eng):
1. G1 - first period grade (numeric: from 0 to 20)
2. G2 - second period grade (numeric: from 0 to 20)
3. G3 - final grade (numeric: from 0 to 20, output target)

U korištenom datasetu nalazi se 395 učenika.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mat = pd.read_csv("student-mat.csv")
print(mat.head)
```

## Obrada podataka

Prvi graf prikazuje podjelu po spolu. Prevladava ženski spol. Djevojaka ima 208, dok je mladića 187.

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.countplot(x="sex", data=mat)
mat["sex"].value_counts()
```

Drugi graf pokazuje raspodjelu učenika po godinama. Kao što se vidi radi se o učenicima u dobnoj skupini od 15 do 19 godina, uz iznimke nekoliko ponavljača.

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.countplot(x="age", data=mat)
plt.title('Students distribution by age, histogram');
print("Prosječna dob učenika je", mat["age"].mean())
```

Učenici koji se nalaze u datasetu pohađaju dvije različite škole. U heatmapi ispod pokazan je broj pojedinih učenika koji pohađaju gimnaziju Gabriel Pereira (GP) i gimnaziju Mousinho da Silveira (MS) u ovisnosti o spolu. Prevladavaju učenici gimnazije Gabriel Pereira.

```python
sns.heatmap(pd.crosstab(mat['sex'], mat['school']), annot=True, fmt='d', cmap="Blues")
gp = len(mat[mat['school']=='GP'])
ms = len(mat[mat['school']=='MS'])
print("U skupini anketiranih učenika, školu Gabriel Pereira pohađa",gp,", a školu Mousinho da Silveira pohađa", ms, "učenika.")
```

U sljedećem dijelu koda spajamo vrijednosti edukacije oca i majke učenika u zajedničku varijablu nazvanu parents_ed. Umjesto varijabli Medu i Fedu koje rangiraju od 0 do 4, nova varijabla rangira od 0 do 8.

```python
#zbroj stupnja edukacije roditelja
mat['parents_ed'] = mat['Medu'] + mat['Fedu']
print(mat['parents_ed'])
```

```python
#prosjek ocjena za matematiku
print(mat['G3'])
```

Prikazan je histogram koji prikazuje distribuciju učenika u odnosu na njihovu prosječnu ocjenu iz matematike. Vidljivo je da je distribucija otprilike pravilna prema sredini, uz blagu prednost prema većim ocjenama. Prosječni broj bodova iznosi 10.41.

```python
sns.histplot(data = mat, x = mat['G3'], kde = True, color = 'Green')
sns.set(rc={'figure.figsize':(14.7,8.27)})
plt.title("Average Mat grades")
plt.ylabel("# of Students")
plt.xlabel("Score / 20")
print(mat['G3'].mean())
```

U prikazanom jointplot grafu prikazana je distribucija učenika u ovisnosti o prosječnom broju bodova predmeta i kombiniranoj razini edukacije roditelja. Vidljiva je pozitivna korelacija između te dvije varijable, dakle veći stupanj edukacije roditelja statistički podrazumijeva i veće ocjene učenika. Faktor korelacije je otprilike 0.222, što označava slabu, ali ipak postojeću korelaciju.

```python
sns.jointplot(mat['parents_ed'], mat['G3'], kind="reg");
print("Korelacija je: ", mat['parents_ed'].corr(mat['G3']))
```

Prikazani graf je histogram koji prikazuje samoocijenjenu količinu napora kojeg učenici ulažu u učenje. Kao što je vidljivo, najveći dio učenika svoj trud ocjenjuju sa 2 od 4.

```python
mat['studytime'].value_counts().plot(kind = 'bar', title = 'Students Studytime (Rated 1-4)', color = 'lightgreen')
```

Boxplot ispod pokazuje distribuciju ocjena u ovisnosti o količini učenja. Vidimo iz grafa da generalno više truda donosi bolje rezultate, ali izlazi i iznenađujući zaključak da postoji i određena granica nakon koje više učenja ne mora nužno značiti i veće ocjene, što se vidi u distribucijama za 3 i 4, koje imaju skoro identičan prosjek.

```python
plt.figure(figsize=(18,7))
plt.title("Box plot konačnih ocjena u ovisnosti o količini učenja")
sns.boxplot(y=mat["studytime"], x="G3", data = mat , orient="h", palette = 'rainbow')
```

U prikazanom histogramu (ispod) prikazana je konzumacija alkohola učenika. Zbrajanjem varijabli Dalc (konzumacija alkohola preko radnog tjedna) i Walc (konzumacija alkohola vikendom) u jednu varijablu DWalc dobijena je ukupna konzumacija alkohola. Kako se količina konzumacije rangira od 1 do 5 (1 predstavlja da učenik ne pije alkohol, 2 jako malo, 3 umjereno, 4 poprilično i 5 jako puno) najmanja moguća vrijednost konzumacije je 2 a najveća 10. Iz histograma se vidi da većina učenika, preko 140 uopće ne konzumira alkohol, dok njih svega nekoliko konzumira alkohol u zabrinjavajućim količinama.  

```python
mat['DWalc'] = mat['Dalc'] + mat['Walc']
```

```python
list = []
i=2
for i in range(2,11):
    list.append(len(mat[mat.DWalc == i]))
ax = sns.barplot(x = [2,3,4,5,6,7,8,9,10], y = list)
plt.ylabel('Number of Students')
plt.xlabel('Weekly alcohol consumption')
```

```python
ne_piju = len(mat[mat['DWalc']==2])
postotak_ne_piju = round(((ne_piju/len(mat['DWalc']))*100),2)
print("Iz ankete je dobijeno da ", ne_piju, "učenika konzumiraju alkohol u zanemarivim količinama, što čini", postotak_ne_piju, "% svih ispitanih.")
```

Kao što je vidljivo iz donjeg grafa, internet čini veliku razliku u uspjehu učenika budući da oni koji imaju pristup internetu mogu koristiti širi spektar resursa za učenje.

```python
plt.figure(figsize=(17,5))
plt.title("Box plot konačnih ocjena u ovisnosti ima li učenik pristup internetu")
sns.boxplot(y=mat['internet'], x="G3", data = mat , orient="h", palette = 'tab10')
```

```python
parents_alc = mat[mat['Pstatus'] == 'A']
p_a = parents_alc['DWalc'].mean()
p_b = mat['DWalc'].mean()
print("separated parents dalc vals " + str(round(p_a, 3)) 
      + " vs. those with nonseparated parents " + str(round(p_b,3)))
```

U sljedećem dijelu razmatrati ćemo potporu obitelji i društva na ocjene i stopu konzumacije alkohola. Kao kombinirani faktor, pretvorit ćemo socijalne parametre koji imaju odgovor sa da/ne u težinske faktore: 0 za ne, i 1 za da. Nakon toga ćemo ih zbrojiti u općenitu ocjenu potpore.
Kao što je vidljivo iz prvog grafa, potpora obitelji i škole svakako pomaže, ali previše potpore zna imati i suprotan efekt.
Iz drugog grafa vidljivo je da povećanjem potpore obitelji i škole pada stope konzumacije alkohola.

```python
mat.loc[mat["schoolsup"] == "yes", "schoolsup"] = 1
mat.loc[mat["schoolsup"] == "no", "schoolsup"] = 0

mat.loc[mat["famsup"] == "yes", "famsup"] = 1
mat.loc[mat["famsup"] == "no", "famsup"] = 0

mat.loc[mat["paid"] == "yes", "paid"] = 1
mat.loc[mat["paid"] == "no", "paid"] = 0

mat.loc[mat["nursery"] == "yes", "nursery"] = 1
mat.loc[mat["nursery"] == "no", "nursery"] = 0

mat["socscore"] = mat["schoolsup"] + mat["famsup"] + mat["paid"] + mat["nursery"]

plt.figure(figsize=(17,5))
plt.title("Box plot konačnih ocjena u ovisnosti o faktorima potpore škole i obitelji")
sns.boxplot(y=mat['socscore'], x="G3", data = mat , orient="h", palette = 'tab10')

plt.figure(figsize=(17,5))
plt.title("Box plot stopa konzumacije alkohola u ovisnosti o faktorima potpore škole i obitelji")
sns.boxplot(y=mat['socscore'], x="DWalc", data = mat , orient="h", palette = 'tab10')
```

```python
average=mat['G3'].mean()
mat['average'] = ['under average'if i < average else 'above average' for i in mat.G3]
sns.swarmplot(x='DWalc',y='G3',hue='average',data=mat,palette={'above average':'Red','under average':'green'})
```

Iz navedenog grafa vidljivo je da nema povezanosti između stope konzumacije alkohola i izostanaka učenika iz škole.

```python
sns.swarmplot(x='DWalc',y='absences',hue='average',data=mat,palette={'above average':'Red','under average':'green'})
print("Faktor korelacije izostanaka i stope konzumacije alkohola je", round(mat["DWalc"].corr(mat["absences"]),2), ",što nije značajna vrijednost")
```

Iz navedenog grafa vidljivo je da učenici koji imaju više slobodnog vremena su ujedno i skloniji alkoholu, iz čega se može zaključiti da učenici koji preokupiraju svoje vrijeme s više obaveza imaju manje prilika doći u napast alkohola.

```python
plt.figure(figsize=(17,5))
plt.title("Box plot stope konzumacije alkohola u ovisnosti o količini slobodnog vremena")
sns.boxplot(y="freetime", x="DWalc", data = mat , orient="h", palette = 'tab10')
```

Iz donjeg grafa vidljivo je da u dosta slučajeva gori odnosi u obitelji znače da će dijete biti sklonije alkoholu, iako postoje iznimke u svakoj kategoriji. Korelacija međutim nije velika.

```python
plt.figure(figsize=(17,5))
plt.title("Box plot stopa alkoholizma u ovisnosti o odnosima u obitelji")
sns.boxplot(y="famrel", x="DWalc", data = mat , orient="h", palette = 'tab10')

print("Faktor korelacije stope konzumacije alkohola sa odnosima u obitelji je", round(mat["DWalc"].corr(mat["famrel"]),2))
```

Donji graf prikazuje matricu korelacija svake od varijabli. Matrica je zrcalna s obzirom na glavnu dijagonalu.

```python
plt.figure(figsize=(15,15))
sns.heatmap(mat.corr(),annot= True, fmt = ".2f", cbar = True, cmap='Blues')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
```
